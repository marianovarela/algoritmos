```python
import random
import timeit
import matplotlib.pyplot as plt
import numpy as np

def insertionSort(lista):
    for i in range(1,len(lista)):
        #print(i)
        #print(lista[i])
        current = lista[i]
        #print("empieza while")
        while i>0 and lista[i-1]>current:
            #print("dentro del while")
            #print(i)
            lista[i] = lista[i-1]
            i = i-1
            lista[i] = current
    return lista

n_min = 10
n_max = 1000
st = 10
iteraciones = 10 # iteraciones del benchmark

t_insert_1 = list()
for n in range(n_min, n_max, st):
    suma = 0
    for i in range(1, iteraciones):
        v = random.sample(range(1, 1000000), n)
        suma += timeit.timeit(lambda: insertionSort(v), number = 1)
    promedio = suma / iteraciones
    t_insert_1.append(promedio)
    
eje_x = list(range(n_min, n_max, st))
plt.xlabel('n')
plt.ylabel('costo')


plt.plot(eje_x, t_insert_1, label = 'insertion')
plt.legend()
plt.show()
    
```
